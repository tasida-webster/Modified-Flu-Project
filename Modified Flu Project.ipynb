# Modified Flu Project

### Overview

We have been given the task of adjusting the original SIR model presented in chapter 4 section 3 of Shiflets’ Introduction to Computational Sciences. The adjustments made to the model will allow us to analyze the spread of influenza in a closed environment, a boys’ boarding school. We began by examining each component of the SIR model--susceptibles, infecteds, and recovereds--to gain a clear understanding of how the model works without any vaccinations given to the boys. Once we understood its implications, we then added different variations to the original SIR model that simulate real world scenarios in context to the influenza outbreak. More specifically, we tested an increase in recovery time, the implementation of social distancing, and the effects of people not following social distancing. This paper will analyze in more detail the methods we used.

## Original Model

```python
DT = 0.25 # Our time step was a quarter of a day
simLength = 14 #We will run the code for 14 days
numIterations = int(simLength/DT) + 1 #The number iterations the for loop will run for.
t = 0 #time in days

susceptibles = 762 #Boys
infecteds = 1 #Boys
recovereds = 0 #Boys
vaccinated=0 #Boys
totalwhogotsick=infecteds


infection_rate = 0.00218#The number of human intreaction the boys incounter(33.3) times the spread of the disease(5%)
#divided by the total number of boys(763)
recovery_rate = 0.5 #A boy is usually sick for 2 days so 50% of the infected boys get well each day

get_sick = infection_rate * susceptibles * infecteds #Determines who gets infected 
recover = recovery_rate * infecteds #50% of the infecteds recover

tLst = [t]
SLst = [susceptibles]
ILst = [infecteds]
RLst = [recovereds]
totalwhogotsicklist=[totalwhogotsick]

for i in range(1, numIterations):
    t = i * DT #Calculating the time
    susceptibles = susceptibles + (-get_sick) * DT #Subtracting the number of boys that get infected 
    #from the susceptibles.
    infecteds = infecteds + (get_sick - recover) * DT #Calculating how many boys get infected
    recovereds = recovereds + (recover) * DT #Calculating how many boys recover
    totalwhogotsick=totalwhogotsick +(get_sick) * DT

    get_sick = infection_rate * susceptibles * infecteds
    recover = recovery_rate * infecteds

    tLst.append(t)
    SLst.append(susceptibles)
    ILst.append(infecteds)
    RLst.append(recovereds)
    totalwhogotsicklist.append(totalwhogotsick)
print(totalwhogotsick)
```
